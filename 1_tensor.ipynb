## Einsum   .. Einstein summation notation

```python
        ## transpose the tensor .. (change the shape)
x = torch.rand(3, 4)
# x = torch.rand(3, 4, 5)
print(x)

z = torch.einsum("ij -> ji", x)            ## i => matrix shape[0]         j => matix shape[1]
# z = torch.einsum("ijk -> kji", x)             ## i => matrix shape[0]         j => matix shape[1]          k => matix shape[2]
print(z)
```

```python
        ## summation
x = torch.rand(3, 4)
print(x)
z = torch.einsum('ij -> ', x)       ## it will add all numbers present in the tensor and return the sum value
print(z)
```

```python
        ## column summation
x = torch.rand(3, 4)
print(x)
z = torch.einsum('ij -> j', x)  ## it will add numbers present in the individual columns and return total values of all individual columns
print(z)
```

```python
        ## row summation
x = torch.rand(3, 4)
print(x)
z = torch.einsum('ij -> i', x)  ## it will add numbers present in the individual rows and return total values of all individual rows
print(z)
```

```python
        ## element wise multiplication
x = torch.rand(3, 4)
y1 = torch.rand(4, 3)
y2 = torch.rand(2, 4)

print(x)
# print(y1)
# print(y2)
z1 = torch.einsum("ij,jk -> ik", x,y1)      ## i => matrix1 and matrix2 shape[0]         j => matix1 shape[1]       k => matix2 shape[1]
                                            ## matrix1 column and matrix2 row will multiply and form a new resultant value of shape (i,k)
z2 = torch.einsum("ij,kj -> ik", x,y2)          ## matrix1 column and matrix2 column will multiply and form a new resultant value
print(z1)
print(z2)
```

```python
        ## dot product with matrix

x = torch.rand((2,3))
y = torch.rand((3,2))

z = torch.einsum('ij, kl ->', x,y)
print(z)
```

```python
x = torch.rand(3, 2)
print(x)

        ## dot product first row with first row of the matrix
# z = torch.einsum("i,i ->", x[0], x[0])
z = torch.einsum("i,j ->", x[0], x[0])
print(z)
```

```python
        ## outer product
x = torch.rand((2))
y = torch.rand((3))

z = torch.einsum('i, j -> ij', x,y)
print(z)
```

```python
        ## batch matrix multiplication
x = torch.rand((2, 3, 5))
y = torch.rand((2, 5, 4))

z = torch.einsum('ijk, ikl -> ijl', x,y)
print(z.shape)
```

### triangual value matrix

```python
n = 4  # Number of rows and columns
tensor = torch.tril(torch.ones(n, n), diagonal=0)

print(tensor)
```

```python
n = 4  # Number of rows and columns

# Create a tensor of ones
tensor = torch.tril(torch.ones(n, n), diagonal=0)

# Replace zeros with -inf
tensor[tensor == 0] = float('-inf')

print(tensor)
```

### masked_fill

```python
energy = torch.rand(2,8,9,9)
mask = torch.rand(2,9).reshape(2,1,1,9)
xx = energy.masked_fill(mask == 0, float("-1e20"))          ## if the mask is not 0 then create a new tensor of dim (2,8,9,9)
                                                            ## and fill random vlaue between ( -1e20 )
# print(xx[0])
print(xx.shape)
```

```python
attention = torch.softmax(xx / (256 ** (1 / 2)), dim=3)
print(attention.shape)
# print(attention[0])
```

## Torch and Numpy

```python
import numpy as np
```

### torch to numpy

```python
x = torch.ones(5)
print(type(x))

y = x.numpy()                       ## both x and y are sharing same memory ..it means if you modify x then y will also change
print(type(y))
```

```python
x.add_(2)
print(x)
print(y)
```

### numpy to torch

```python
x = np.ones(5)
print(x)

y = torch.from_numpy(x)             ## both x and y are sharing same memory ..it means if you modify x then y will also change
print(y)
```

```python
x = x + 1
print(x)
print(y)
```
